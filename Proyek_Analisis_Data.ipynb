Keterangan Nama Variabel

**PM2.5** = PM2.5 mengacu 	pada partikel-partikel kecil atau debu halus dalam udara yang memiliki diameter kurang dari atau sama dengan 2.5 mikrometer (µm).

**PM10** = PM10 mengacu 	pada partikel-partikel kecil atau debu halus dalam udara yang memiliki diameter kurang dari atau sama dengan 10 mikrometer (µm).

**DEWP** = Titik embun (Dew Point)

**WD** = Arah angin (Wind Direction)

**WSPM** = Kecepatan angin (Wind Speed)

### Pertanyaan 3: Bagaimana kualitas udara Station Wanliu dibandingkan dengan Station Tiantan?

````python
# Membuat histogram untuk PM2.5
plt.hist(wanliu_df['PM2.5'], bins=20, alpha=0.5, label='wanliu')
plt.hist(tiantan_df['PM2.5'], bins=20, alpha=0.5, label='tiantan')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Histogram PM2.5 Wanliu dan Tiantan')
plt.legend()
plt.show()

#Membuat histogram untuk PM10
plt.hist(wanliu_df['PM10'], bins=20, alpha=0.5, label='wanliu')
plt.hist(tiantan_df['PM10'], bins=20, alpha=0.5, label='tiantan')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Histogram PM10 Wanliu dan Tiantan')
plt.legend()
plt.show()

#Membuat histogram untuk SO2
plt.hist(wanliu_df['SO2'], bins=20, alpha=0.5, label='wanliu')
plt.hist(tiantan_df['SO2'], bins=20, alpha=0.5, label='tiantan')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Histogram SO2 Wanliu dan Tiantan')
plt.legend()
plt.show()

#Membuat histogram untuk NO2
plt.hist(wanliu_df['NO2'], bins=20, alpha=0.5, label='wanliu')
plt.hist(tiantan_df['NO2'], bins=20, alpha=0.5, label='tiantan')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Histogram NO2 Wanliu dan Tiantan')
plt.legend()
plt.show()

#Membuat histogram untuk CO
plt.hist(wanliu_df['CO'], bins=20, alpha=0.5, label='wanliu')
plt.hist(tiantan_df['CO'], bins=20, alpha=0.5, label='tiantan')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Histogram CO Wanliu dan Tiantan')
plt.legend()
plt.show()

#Membuat histogram untuk O3
plt.hist(wanliu_df['O3'], bins=20, alpha=0.5, label='wanliu')
plt.hist(tiantan_df['O3'], bins=20, alpha=0.5, label='tiantan')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Histogram O3 Wanliu dan Tiantan')
plt.legend()
plt.show()
````
